```python
import os
import random as rn 
from shutil import copy2
```

```python
target_folder = r"C:\Users\agah2\Desktop\JupyterNotebooks\BIP\dataset"
```

```python
# Ratios of images: 80% of images for training etc
train_ratio = 0.8
test_ratio = 0.1
val_ratio = 0.1
```

```python
# Create Training, testing and validation folders
os.makedirs(os.path.join(target_folder, "train_original"), exist_ok=True)
os.makedirs(os.path.join(target_folder, "test_original"), exist_ok=True)
os.makedirs(os.path.join(target_folder, "val_original"), exist_ok=True)
```

```python
# Source
source_folder = r"C:\Users\agah2\Desktop\JupyterNotebooks\BIP\dataset\all_original"
```

```python
# Get list of all image files in source folder
image_files = [f for f in os.listdir(source_folder) if os.path.isfile(os.path.join(source_folder, f))]
```

```python
# Use seed to get reproducible split of images
rn.seed(11)

# Shuffle the list to ensure randomness
rn.shuffle(image_files)
```

```python
# Split images into train, test, and val
train_split = int(len(image_files) * train_ratio)
test_split = int(len(image_files) * test_ratio)
val_split = len(image_files) - train_split - test_split
```

```python
print(f"Total images: {len(image_files)}")
print(f"Train: {train_split}, Test: {test_split}, Val: {val_split}")
```

```python
# Move files to respective folders
for i, file in enumerate(image_files):
    source_path = os.path.join(source_folder, file)
    
    if i < train_split:
        destination_path = os.path.join(target_folder, "train_original", file)
    elif i < train_split + test_split:
        destination_path = os.path.join(target_folder, "test_original", file)
    else:
        destination_path = os.path.join(target_folder, "val_original", file)

    copy2(source_path, destination_path)
```
