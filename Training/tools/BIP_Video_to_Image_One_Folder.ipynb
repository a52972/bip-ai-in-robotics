```python
# Name of target folder
training_images_parent_path = os.path.join(base_path, "BIP_images_roboter_cam_allinone")  
print(training_images_parent_path)
```

```python
# Create folder for all images
os.makedirs('all', exist_ok = True)
```

```python
# Name of target folder
training_images_path = os.path.join(training_images_parent_path, "all")  
print(training_images_path)
```

```python
ffmpeg_path = r"C:\Users\agah2\AppData\Local\Microsoft\WinGet\Links\ffmpeg.exe"
```

```python
# Now loop through videos and create images from each video and save in training images folders

# Loop through all mp4 files in the input folder
for filename in os.listdir(training_videos_path):
    if filename.lower().endswith(".mp4"):
        video_path = os.path.join(training_videos_path, filename)
        video_name = os.path.splitext(filename)[0]   # filename without extension
        
        # Images will go directly into training_images_path
        # Prefix frames with the video name to avoid collisions
        output_pattern = os.path.join(training_images_path, f"{video_name}_%04d.jpg")
        
        cmd = [
            ffmpeg_path,
            "-i", video_path,
            "-vf", "fps=1",       # 2 frames per second = one frame every 0.5 seconds
            output_pattern
        ]

        print(f"Extracting frames from {filename}")
        subprocess.run(cmd, check=True)

print("Done!")
```
